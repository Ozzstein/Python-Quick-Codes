```python
def imperial_to_metric(feet=0, inches=0): # keyword arguments
    total_inches = feet * 12 + inches
    cms = total_inches * 2.54
    return(cms / 100)
```

```python
imperial_to_metric(0, 10)
```

```python
imperial_to_metric(inches=10)
```

```python
def F_to_C(degrees_F, Kelvin_offset=0):
    degrees_C = (degrees_F-32) * (5/9)
    degrees_K = degrees_C + Kelvin_offset
    return degrees_C + degrees_K
```

```python
F_to_C(-10, 273.15)
```

```python
F_to_C(0, 273.15)
```

```python
F_to_C(100, 273.15)
```

### Exercise: convert temperatures from C to F

```python
def centigrade_to_Fehrenheit(degrees):
    return (degrees * (9 / 5)) +32
```

```python
temprature_C = [21, 27, 20, 21, 20, 23, 22, 32]
```

```python
results = []
for t in temprature_C:
    F = centigrade_to_Fehrenheit(t)
    results.append(F)
```
