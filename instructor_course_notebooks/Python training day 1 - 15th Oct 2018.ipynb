### Exercise: Writing your own slightly more complex function

You're going to write a temperature conversion function to convert from fahrenheit to either degrees C or degrees K. The way you do this is to define a function `fahrenheit_to_centigrade` that takes two arguments: one required (which is the degrees F) and an optional offset value which is by defauly zero. In the function subtract the offset from the converted degress centigrade.

Try with the following values:

- -10 F to Centigrade and Kelvin
- 0 to Centigrade and Kelvin
- 100 to Centigrade and Kelvin

The formulas are:

$$
F = \frac{9}{5} * C + 32 \\
C = \frac{F - 32}{\frac{9}{5}} \\
K = C + 273.15
$$

```python
def fahrenheit_to_centigrade(degrees, kelvin_offset=0):
    deg_c = (degrees - 32) * 5 / 9
    return deg_c + kelvin_offset
```

```python
fahrenheit_to_centigrade(0, kelvin_offset=273.15)
```

```python
fahrenheit_to_centigrade(32)
```

```python
fahrenheit_to_centigrade?
```

```python
def fahrenheit_to_centigrade(degrees, kelvin_offset=0):
    """Convert fahreheit to degrees C
    
    Converts fahrenheit using the formula C = (F - 32) * 5 / 9
    
    Optionally you can specify an offset for Kelvin (should be
    273.15)
    
    Examples
    
    >>> fahrenheit_to_centigrade(32)
    0.0
    """
    deg_c = (degrees - 32) * 5 / 9
    return deg_c + kelvin_offset
```

### Exercise: Returning more than one value from a function

Functions in Python can return any number of values. Separate different values with a `,`. e.g.

    rerturn val1, val2, val3, ...
    
You've already seen a imperial to metric function to convert feet and inches to metres. Write the opposite function that converts metres to feet and inches. As a couple of hints:

- In Python 3, `10 / 3` is 3.333333. If you want integer division use `10 // 3`
- Python has a modulus / remainder keyword `%`, e.g. `14 % 12` gives `2`

```python
def metric_to_imperial(metres):
    """Returns (feet, inches) from metres"""
    total_inches = metres * 100 / 2.54
    feet = total_inches // 12
    inches = total_inches % 12
    
    return feet, inches
```
