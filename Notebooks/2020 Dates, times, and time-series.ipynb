Time series
-----------------------

* “A sequence of data points measured typically at successive times”
* Represents the evolution of a dynamic system over time.
* Time series analysis (TSA) aims at identifying, modelling, and forecasting temporal patterns in time series data.
* Effective TSA needs a class whose data are indexed by times or dates.

Dates and Times
-----------------------
First, you will learn about the basics of the *datetime* package, a built-in package of Python for date and time handling.

```
from datetime import date, time, datetime, timedelta
```

```
# The date class is for a date.
# date(year, month, day)
date(2012,1,1)
```

```
# Check a leap year.
date(2011,2,29)
```

```
# The time class is for a time.
# time(hour, min, second, microsecond)
time(12,1,1,10)
```

```
# The datetime class is for a date with time.
datetime(2012,1,1,12,1,1,10)
```

```
# timedelta is about a difference between two dates or datetimes.
# Question: how many days in the year 2012?
date(2012,1,1) - date(2011,1,1) # 365 days
```

```
# What is the date that is 365 days after 1st Jan 2011?
date(2011,1,1)+timedelta(days=365)
```

```
# Combine a date and a time into a datetime object.
datetime.combine(date(2012,12,31), time(1,1,1))
```

```
# Modify a date.
date(2012,12,31).replace(day=10)
```

The *datetime* package is good for handling an individual date and/or time.

Date range
-----------------------
The *Pandas* package provides powerfull utilities to generate a range of dates. We will use these utilities when indexing a time series later.

```
import pandas as pd
```

```
# Generate business days between two dates.
pd.bdate_range("2000-01-01", "2000-01-31")
```

```
# When you generate date ranges, you will use the following alias names.
from IPython.core.display import Image # To display images.
Image(filename="extras/date_offset.png")
```

```
# Show me business days in the year 2000.
pd.date_range("2000-01-01", "2000-12-31", freq="BM")
```

```
# Check if the last business day in 2000 is Friday.
date(2000,12,29).weekday() # 0 = Monday, ..., 5 = Friday, ...
```
